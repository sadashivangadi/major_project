```python
import pandas as pd 
import matplotlib.pyplot as plt

df = pd.read_csv("Housing.csv")
# Convert rupees â†’ lakhs
df['price_lakhs'] = df['price'] / 100000

avg_price_ac=df.groupby('airconditioning')['price_lakhs'].mean()
avg_price_ac
```

```python
plt.figure(figsize=(6,4))
plt.bar(avg_price_ac.index, avg_price_ac.values, color=['orange', 'green'])
plt.xlabel("Air Conditioning")
plt.ylabel("Average Price (Lakhs)")
plt.title("Average House Price: Ac / Non Ac Houses")
plt.show()
```
