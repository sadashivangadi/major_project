```python
import pandas as pd
import matplotlib.pyplot as plt     
df = pd.read_csv("Housing.csv")
# Convert rupees â†’ lakhs
df['price_lakhs'] = df['price'] / 100000

parking_price=df.groupby('parking')['price_lakhs'].mean()
parking_price
```

```python
plt.figure(figsize=(6,4))
plt.bar(parking_price.index.astype(str), parking_price.values, color=['purple', 'cyan'])
plt.xlabel("Parking Available")
plt.ylabel("Average Price (Lakhs)")
plt.title("Average House Price: Parking availability diveded in slots like 0 and 1,")
plt.grid(axis='y', linestyle='-', alpha=0.4, linewidth=1.7, color='black')
plt.show()  
```
