```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Housing.csv")

# Convert price into lakhs
df['price_lakhs'] = df['price'] / 100000
# Clean 'prefarea' column
df["prefarea"] = df["prefarea"].str.strip().str.title()

# Category A: <5000 sqft AND no preface
cat_A = df[(df["area"] < 5000) & (df["prefarea"] == "No")]

# Category B: >=5000 sqft AND preface available
cat_B = df[(df["area"] >= 5000) & (df["prefarea"] == "Yes")]



avg_A = cat_A["price_lakhs"].mean()
avg_B = cat_B["price_lakhs"].mean()

price_gap = avg_B - avg_A

print("Average Price (<5000 sqft & No Prefarea):", avg_A)
print("Average Price (>=5000 sqft & Prefarea):", avg_B)
print("Price Gap:", price_gap)
```

```python
plt.figure(figsize=(6,4))
labels = ['<5000 sqft & No Prefarea', '>=5000 sqft & Prefarea', 'Price Gap']
values = [avg_A, avg_B , price_gap]
plt.bar(labels, values, color=['red', 'blue', 'green'])#bar plot with different colors
plt.ylabel("Average Price (Lakhs)"),plt.xlabel("Categories")#labeling x and y axis
plt.title("Average House Price Comparison")#title of the plot
plt.grid(axis='y', linestyle='--', alpha=0.7)#grid lines in y axis
plt.xticks(rotation=15)#showinng. thevaluein x axis with rotation of 15 degree
plt.show()
```
